```python
import numpy as np
import torch 
from torch.autograd import backward

# Input (temp, rainfall, humidity)
inputs = np.array([[73, 67, 43],
                   [91, 88, 64],
                   [87, 134, 58],
                   [102, 43, 37],
                   [69, 96, 70]], dtype='float32')

# Target (apples)
targets = np.array([[56],
                    [81],
                    [119],
                    [22],
                    [103]], dtype='float32')

x_input = torch.tensor(inputs)
x_target = torch.tensor(targets)
print(x_input)
print(x_target)

```

```python
weight = torch.randin(3,1,requires_grade = True)
bias = torch.randin(1,1,requires_grade = True)

print(weight)
print(bias)
```

```python
mul = np.dot(x_input, weight)
tensor_mul = torch.tensor(mul)
model = tensor_mul + bias
print(model)
```

```python
comparison = model - targets
print(comparison)
square = comparison * comparison
def MSE(square):
  sum = 0
  for i in range(len(square)):
    sum += square[i]
  print(sum)
  print((sum)/(2*(len(square))))
MSE(square)  
```
