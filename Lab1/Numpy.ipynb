```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('https://raw.githubusercontent.com/ShuklaRishi/CE147_ML/main/Lab1/Iris.csv')
str1=["1","2","3","4","5"]
X=data['SepalLengthCm']
Y=data['SepalWidthCm']
plt.scatter(X,Y,s=10,color="grey")
plt.xlabel("X")
plt.ylabel("Y")
for i, label in enumerate(str1):
    plt.annotate(label, (X[i], Y[i]))
plt.show()
```
